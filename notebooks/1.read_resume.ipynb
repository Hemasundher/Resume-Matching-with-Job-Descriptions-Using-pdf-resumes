```python
# import PyPDF2
import pandas as pd
```

```python
#creating a df to store all data related of pdf
col_names=['id','number_of_pages','text','category']
main_df=pd.DataFrame(columns=col_names)
main_df
```

```python
count=0
for category,files in pdf_path_dic.items():
    for file_id,path in files:
        pdf_file=open(path,'rb')
        pdf_reader=PyPDF2.PdfReader(pdf_file)

        if pdf_reader.is_encrypted:
            print(f'{id }this file is encrypted')
            continue

        no_of_pages=len(pdf_reader.pages)
        text=''

        for  i in range(no_of_pages):
            page=pdf_reader.pages[i]
            text+="\n"+page.extract_text()

        #creating a new df then concat with old, update the old df
        new_row_df=pd.DataFrame({'id':int(file_id),'number_of_pages':no_of_pages,'text':text,'category':category},index=[0])
        main_df=pd.concat([main_df,new_row_df],axis=0,ignore_index=True)


        count+=1


```

preprocess the text

*   clean the text
*   generate tokens



```python
import spacy
nlp = spacy.load("en_core_web_lg")
```

```python
import re
```

```python

def preprocess(text):
    # convert all characters in the string to lower case
    text = re.sub(r'\n', ' ', text)

    #remove digits
    text = re.sub(r'\d', '', text)

    # Remove punctuation (you can customize this pattern for your needs)
    text = re.sub(r'[^\w\s]', '', text)

    # Remove extra spaces and replace with a single space
    text = re.sub(r'\s+', ' ', text).strip()

    # tokenize word
    doc=nlp(text)
    toks=[i.text.lower() for i in doc if not i.is_stop]


    return " " .join(toks)
```
