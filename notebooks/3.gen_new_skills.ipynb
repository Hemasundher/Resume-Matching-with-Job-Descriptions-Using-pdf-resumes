reading json files , getting skills from it

```python
import json
```

```python
def read_json(path):
    output_data=[]
    with open(path, "r") as file:
        for line in file:
            data = json.loads(line)
            output_data.append(data)
    return output_data
```

```python
def json_to_skills(path):

    json_list=read_json(path)
#     print(json_list)

    n=len(json_list)
    all_skills=set()
    s=[]
    for x in range(n):
        for i in range(len(json_list[x]['pattern'])):
            try:
                s.append(json_list[x]['pattern'][i]['LOWER'])
            except:
                s.append(json_list[x]['pattern'][i]['TEXT'].lower())


        string=' '.join(s)
        s=[]
        all_skills.add(string)
    return all_skills
```

```python
main_json_skills=json_to_skills('jz_skill_patterns.jsonl')
```

*   generate new skills from spacy trained model




```python
def generate_new_skills(probability_threshold=0.87):
    cat=list(set(df['category']))
    new_gen_skills=[]
    for category in cat:
        all_skills=set()
        for i in range(len(df)):
            if df.loc[i]['category']=='ACCOUNTANT':
                s=set(df.loc[i]['skills'].split(','))
                all_skills=s.union(all_skills)


        for skill in list(all_skills):
            pred=model.get_nearest_neighbors(skill,k=10)
            new_set=set()
            for new in pred:
                dic={"label":"SKILL","pattern":[]}

                if new[0]<probability_threshold or new[1] in main_json_skills or new[1] in new_set:
                    continue
                new_set.add(new[1])

                for word in new[1].split():

                    dic['pattern'].append({"LOWER":word})

                new_gen_skills.append(dic)

    return new_gen_skills




```

```python
new_gen_skills=(generate_new_skills())
# 0.7155979871749878<0.8
```

writing a json files with newly generated skills

```python
def write_json_with_list(l,path):
    with open(path, "w") as json_file:
    # Write the list of values to the JSON file
     for i in l:
        json.dump(i, json_file)
        json_file.write('\n')
```

```python
#path to new jsonl file to add newly gen skills
path='/content/drive/MyDrive/Colab Notebooks/resume skills - assessment/JsonFiles/new_gen_skill.jsonl'

write_json_with_list(new_gen_skills,path)
```

combining two json files

```python
def merge_two_jsons(path1,path2,new_path):
    l1=read_json(path1)
    l2=read_json(path2)
    l=l1+l2

    write_json_with_list(l,new_path)


```
